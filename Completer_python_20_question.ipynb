```python
def filter_int(a):
    result = list(filter(lambda x: type(x) == int, a))
    return result

data = [10, "Hello", 20.5, 30, True, "Python", 40]

print(filter_int(data))
```

```python
def filter_int(*args):
    result = list(filter(lambda x: type(x) == int, args))
    return result

print(filter_int(10, "Hello", 20.5, 30, True, 40))
```
